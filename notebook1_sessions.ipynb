```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vrgames_sessions.csv')
print('Rows loaded:', len(df))
```

### Mean Values — Gaming & Platform Variables

```python
print('Variable 1 — Avg Session Duration      :', round(df['session_min'].mean(), 2), 'minutes')
print('Variable 2 — Avg Engagement Rate       :', round(df['engagement_rate'].mean(), 4))
print('Variable 3 — Avg Game Completion Rate  :', round(df['game_completion'].mean(), 4))
print('Variable 4 — Avg FPS                   :', round(df['fps'].mean(), 2), 'fps')
print('Variable 5 — Lag Events Count          :', df['lag_event'].sum(), 'out of', len(df))
```

### Graph — Session Duration Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df['session_min'], bins=30, color='steelblue', edgecolor='white')
plt.axvline(df['session_min'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["session_min"].mean():.1f} min')
plt.axvline(30, color='orange', linestyle='--', label='Target: >30 min')
plt.title('Session Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Engagement Rate Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df['engagement_rate'], bins=25, color='coral', edgecolor='white', alpha=0.85)
plt.axvline(df['engagement_rate'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["engagement_rate"].mean():.3f}')
plt.axvline(0.7, color='orange', linestyle='--', label='Target: 0.70')
plt.title('User Engagement Rate Distribution')
plt.xlabel('Engagement Rate (0-1)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Game Completion Rate by Genre

```python
genre_completion = df.groupby('genre')['game_completion'].mean().sort_values(ascending=False)
plt.figure(figsize=(7, 4))
genre_completion.plot(kind='bar', color='mediumseagreen', edgecolor='white')
plt.axhline(df['game_completion'].mean(), color='red', linestyle='--',
            label=f'Overall Mean: {df["game_completion"].mean():.3f}')
plt.title('Average Game Completion Rate by Genre')
plt.xlabel('Genre')
plt.ylabel('Completion Rate')
plt.xticks(rotation=15)
plt.legend()
plt.tight_layout()
plt.show()
```
