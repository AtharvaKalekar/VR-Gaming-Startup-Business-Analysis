```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vrgames_monitoring.csv')
print('Rows loaded:', len(df))
```

### Mean Values — System Monitoring Variables

```python
print('Variable 6 — Average System Uptime     :', round(df['uptime_pct'].mean(), 4), '%')
print('           — Average FPS               :', round(df['avg_fps'].mean(), 2), 'fps')
print('           — Avg Crash Reports/Day     :', round(df['crash_reports'].mean(), 1))
print('           — Avg Support Tickets/Day   :', round(df['support_tickets'].mean(), 1))
print('ALERT days (uptime < 99 %)             :', df[df['status']=='ALERT'].shape[0], 'out of 365')
```

### Graph — System Uptime over 365 Days

```python
plt.figure(figsize=(10, 4))
plt.plot(df['uptime_pct'], color='steelblue', linewidth=0.9)
plt.axhline(df['uptime_pct'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["uptime_pct"].mean():.4f}%')
plt.axhline(99.9, color='orange', linestyle='--', label='Target: 99.9%')
plt.title('Platform Uptime — 365 Days')
plt.xlabel('Day')
plt.ylabel('Uptime %')
plt.ylim(97.5, 100.1)
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Average FPS over 365 Days

```python
plt.figure(figsize=(10, 4))
plt.plot(df['avg_fps'], color='darkorange', linewidth=0.9)
plt.axhline(df['avg_fps'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["avg_fps"].mean():.2f} fps')
plt.axhline(60, color='green', linestyle='--', label='Target: 60 fps')
plt.title('Average FPS — 365 Days')
plt.xlabel('Day')
plt.ylabel('FPS')
plt.legend()
plt.tight_layout()
plt.show()
```

### Graph — Daily Crash Reports Distribution

```python
plt.figure(figsize=(8, 4))
plt.hist(df['crash_reports'], bins=20, color='tomato', edgecolor='white', alpha=0.85)
plt.axvline(df['crash_reports'].mean(), color='red', linestyle='--',
            label=f'Mean: {df["crash_reports"].mean():.1f}/day')
plt.axvline(10, color='darkred', linestyle='--', label='Alert threshold: 10/day')
plt.title('Daily Crash Reports Distribution')
plt.xlabel('Crash Reports')
plt.ylabel('Number of Days')
plt.legend()
plt.tight_layout()
plt.show()
```
